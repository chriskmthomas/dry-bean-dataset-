```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel ('drybean.xlsx')
df.head(20)
```

```python
df['MajorAxisLength'].fillna(df['MajorAxisLength'].mode()[0], inplace=True)
df['EquivDiameter'].fillna(df['EquivDiameter'].mode()[0], inplace=True)
```

```python
df['roundness'].fillna(df['roundness'].mode()[0], inplace=True)
df['Area'].fillna(df['Area'].mode()[0], inplace=True)
```

```python
df[df==np.inf]=np.nan
df.fillna(df.mean(), inplace=True)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df['Class']= label_encoder.fit_transform(df['Class'])
  
df['Class'].unique()
```

```python
feature_cols = ['Area', 'Perimeter', 'MajorAxisLength', 'MinorAxisLength',
       'AspectRation', 'Eccentricity', 'ConvexArea', 'EquivDiameter', 'Extent',
       'Solidity', 'roundness', 'Compactness', 'ShapeFactor1', 'ShapeFactor2',
       'ShapeFactor3', 'ShapeFactor4']
```

```python
X = df[feature_cols] # Features
y = df['Class']
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
## predict the model
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn import metrics
```

```python

print(confusion_matrix(y_test, y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

```

```python
print(classification_report(y_test, y_pred))
```
